**Model Definition**

Defining the **Transformer** for Claim Verification **(BERT)**

```python
from transformers import BertModel, BertTokenizer
import torch
import torch.nn as nn
import torch.nn.functional as F

# Tokenizer Initialization
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Model Definition
class ClaimVerificationModel(nn.Module):
    def __init__(self, transformer_model_name='bert-base-uncased'):
        super(ClaimVerificationModel, self).__init__()
        # Load a pre-trained BERT model
        self.transformer = BertModel.from_pretrained(transformer_model_name)
        self.dropout = nn.Dropout(0.3)
        self.fc_claim_verification = nn.Linear(self.transformer.config.hidden_size, 2)  # For claim verification
        self.fc_similarity = nn.Linear(self.transformer.config.hidden_size, 128)  # Embedding for similarity

    def forward(self, input_ids, attention_mask, token_type_ids=None, labels=None, mode='verification'):
        # Pass through the transformer
        transformer_output = self.transformer(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids
        )
        cls_output = transformer_output.pooler_output  # [CLS] token embedding

        if mode == 'verification':
            # For claim verification (classification task)
            logits = self.fc_claim_verification(self.dropout(cls_output))
            return logits
        elif mode == 'similarity':
            # For similarity comparison (embedding task)
            embedding = self.fc_similarity(self.dropout(cls_output))
            return embedding
        else:
            raise ValueError("Invalid mode. Choose 'verification' or 'similarity'.")

# Example of Usage
model = ClaimVerificationModel()
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertForSequenceClassification, BertTokenizer, AdamW
import torch.nn as nn
from torch.nn.utils import clip_grad_norm_
from sklearn.metrics import accuracy_score
from tqdm import tqdm
import os

# 1. Define Model Architecture
class ClaimVerificationModel(nn.Module):
    def __init__(self, base_model):
        super(ClaimVerificationModel, self).__init__()
        self.transformer = base_model.bert  # Extract BERT model
        self.dropout = nn.Dropout(p=0.3)
        self.fc_claim_verification = nn.Linear(base_model.config.hidden_size, 2)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.transformer(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        dropped_out = self.dropout(pooled_output)
        logits = self.fc_claim_verification(dropped_out)

        if labels is not None:
            loss_fn = nn.CrossEntropyLoss()
            loss = loss_fn(logits, labels)
            return {'loss': loss, 'logits': logits}
        return {'logits': logits}

# 2. Dataset Class
class ClaimVerificationDataset(Dataset):
    def __init__(self, data, tokenizer, max_len=512):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, item):
        row = self.data.iloc[item]
        label = row['label']
        combined_input = row['title'] + " " + row['text']
        encoding = self.tokenizer.encode_plus(
            combined_input,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# 3. Tokenizer and Model Initialization
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
base_model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
model = ClaimVerificationModel(base_model)


# 5. Create Datasets and DataLoaders
train_dataset = ClaimVerificationDataset(train_data, tokenizer)
val_dataset = ClaimVerificationDataset(val_data, tokenizer)
test_dataset = ClaimVerificationDataset(test_data, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)
test_loader = DataLoader(test_dataset, batch_size=16)

# 6. Training Setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)
num_epochs = 11



# Create a directory for checkpoints
checkpoint_dir = '/content/drive/MyDrive/checkpoints'
os.makedirs(checkpoint_dir, exist_ok=True)

# 7. Training and Evaluation Loop

for epoch in range(num_epochs):
    # Training Phase
    model.train()
    total_loss = 0
    train_preds, train_labels = [], []

    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Training"):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs['loss']
        total_loss += loss.item()


        # Backward Pass
        loss.backward()
        clip_grad_norm_(model.parameters(), max_norm=1.0)  # Gradient clipping
        optimizer.step()

        # Predictions for Accuracy
        logits = outputs['logits'].detach().cpu()
        train_preds.extend(torch.argmax(logits, dim=1).numpy())
        train_labels.extend(labels.cpu().numpy())

    avg_train_loss = total_loss / len(train_loader)
    train_acc = accuracy_score(train_labels, train_preds)
    print(f"Epoch {epoch+1} - Train Loss: {avg_train_loss:.4f}, Train Accuracy: {train_acc:.4f}")


    # Save Checkpoint after Each Epoch
    checkpoint_path = os.path.join(checkpoint_dir, f'epoch_{epoch+1}.pth')
    torch.save(model.state_dict(), checkpoint_path)
    print(f"Checkpoint saved at {checkpoint_path}")

    # Validation Phase
    model.eval()
    total_val_loss = 0
    val_preds, val_labels = [], []

    with torch.no_grad():
        for batch in tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Validation"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs['loss']
            total_val_loss += loss.item()

            # Predictions for Accuracy
            logits = outputs['logits'].detach().cpu()
            val_preds.extend(torch.argmax(logits, dim=1).numpy())
            val_labels.extend(labels.cpu().numpy())

    avg_val_loss = total_val_loss / len(val_loader)
    val_acc = accuracy_score(val_labels, val_preds)
    print(f"Epoch {epoch+1} - Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {val_acc:.4f}")

# 8. Save Final Model
model_save_path = '/content/drive/MyDrive/claim_verification_model_bert.pth'
torch.save(model.state_dict(), model_save_path)
print(f"Model saved to: {model_save_path}")
```

```python
from rouge_score import rouge_scorer # Import the rouge_scorer module from rouge_score
import matplotlib.pyplot as plt
import torch

scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True) # Create a RougeScorer object
rouge_scores = []

# Switch model to evaluation mode
model.eval()

# Loop through test data to calculate predictions and references
with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels']

        # Generate predictions from the model
        outputs = model(input_ids, attention_mask=attention_mask)

        # Access logits from the dictionary
        predictions = torch.argmax(outputs['logits'], dim=1)

        # Convert predictions and references to text format
        for pred, ref in zip(predictions.cpu().numpy(), labels.numpy()):
            predicted_sentence = tokenizer.decode(pred, skip_special_tokens=True)  # Decode predicted tokens
            reference_sentence = tokenizer.decode(ref, skip_special_tokens=True)  # Decode reference tokens

            # Compute ROUGE score using rouge_scorer
            scores = scorer.score(reference_sentence, predicted_sentence)
            rouge_scores.append(scores['rouge1'].fmeasure)  # Using ROUGE-1 F1 score

# Plotting the ROUGE scores
plt.plot(range(len(rouge_scores)), rouge_scores, label="ROUGE-1 F1 Score")
plt.xlabel("Samples")
plt.ylabel("ROUGE Score")
plt.title("ROUGE Score Line Chart")
plt.legend()
plt.show()
```
